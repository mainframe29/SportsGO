```python
import cv2
import numpy as np

from google.colab.patches import cv2_imshow

circle = cv2.imread('shapes/train/circles/drawing(18).png')

cv2_imshow(circle)
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)
```

```python
batch_size = 16

train_generator = train_datagen.flow_from_directory(
        'shapes/train',
        target_size=(28, 28),
        batch_size=batch_size,
        class_mode='categorical',
        subset='training')

test_generator = test_datagen.flow_from_directory(
        'shapes/test',
        target_size=(28, 28),
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation')
```

```python
import tensorflow as tf

def create_network():

    net = tf.keras.models.Sequential([
      tf.keras.layers.Conv2D(filters=6, kernel_size=5, padding='same'),
      tf.keras.layers.AvgPool2D(pool_size=2, strides=2),
      tf.keras.layers.Conv2D(filters=16, kernel_size=5),
      tf.keras.layers.AvgPool2D(pool_size=2, strides=2),
      tf.keras.layers.Flatten(),
      tf.keras.layers.Dropout(0.3),
      tf.keras.layers.Dense(60, activation='sigmoid'),
      tf.keras.layers.Dense(3, activation='softmax')]
              )
    
    return net
rede = create_network()

rede.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
rede.fit(train_generator, epochs = 50, validation_data = test_generator)
```
